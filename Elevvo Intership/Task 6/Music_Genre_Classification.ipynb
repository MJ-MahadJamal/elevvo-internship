```python
# -Task 6: Music Genre Classification (GTZAN) — MFCC + Spectrogram CNN
# - Dataset via kagglehub (dynamic path detection)
# - Approach A: MFCC (tabular) -> RandomForest
# - Approach B: Mel-spectrogram (image) -> MobileNetV2 transfer learning
# - Metrics: Accuracy, classification report, confusion matrix
# - Options: FAST_MODE, model saving to Google Drive

# %% [code]
!pip install -q kagglehub librosa==0.10.1 soundfile scikit-learn tensorflow matplotlib joblib

import os, glob, pathlib, random, itertools
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.ensemble import RandomForestClassifier
import joblib
import kagglehub
import librosa, librosa.display
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.utils import to_categorical

print("TensorFlow:", tf.__version__)

# Config
# Toggles
RUN_TABULAR = True
RUN_IMAGE   = True

# Runtime controls
FAST_MODE = True        # use subset per genre to speed up first run
FILES_PER_GENRE = 80    # GTZAN has ~100/genre; for full run set FAST_MODE=False

# Audio params
SAMPLE_RATE = 22050
DURATION    = 30.0      # seconds (GTZAN clips are ~30s)
MONO        = True

# MFCC features
N_MFCC = 20

# Image model params
IMG_SIZE   = 224
BATCH_SIZE = 16
EPOCHS     = 6 if FAST_MODE else 12
TRANSFER_LEARNING = True  # use MobileNetV2

# Saving
ASK_SAVE_TO_DRIVE = True

# Reproducibility
SEED = 42
random.seed(SEED); np.random.seed(SEED); tf.random.set_seed(SEED)

#  Download & Detect Paths
# Try common GTZAN slugs (some mirrors differ)
POSSIBLE_DATASETS = [
    "andradaolteanu/gtzan-dataset-music-genre-classification",
    "misrakahmed/gtzan-dataset-music-genre-classification"
]

dataset_path = None
for slug in POSSIBLE_DATASETS:
    try:
        p = kagglehub.dataset_download(slug)
        if os.path.exists(p):
            dataset_path = p
            print("Downloaded:", slug, "->", p)
            break
    except Exception as e:
        print("Tried:", slug, "->", e)

if dataset_path is None:
    raise RuntimeError("Could not download GTZAN dataset via kagglehub. Check Kaggle credentials or slug.")

# Find WAV files (common layouts)
cand_patterns = [
    os.path.join(dataset_path, "**", "genres_original", "*", "*.wav"),
    os.path.join(dataset_path, "**", "Data", "genres_original", "*", "*.wav"),
    os.path.join(dataset_path, "**", "GTZAN", "genres_original", "*", "*.wav"),
]
wav_files = []
for pat in cand_patterns:
    wav_files.extend(glob.glob(pat, recursive=True))

if len(wav_files) == 0:
    raise RuntimeError("No .wav files found. Expected '.../genres_original/<genre>/*.wav' somewhere under the dataset folder.")

def infer_genre_from_path(p):
    parts = pathlib.Path(p).parts
    if "genres_original" in parts:
        gi = parts.index("genres_original")
        if gi + 1 < len(parts):
            return parts[gi + 1]
    return pathlib.Path(p).parent.name

records = [(p, infer_genre_from_path(p)) for p in sorted(wav_files)]
df_all = pd.DataFrame(records, columns=["path", "genre"])
genres = sorted(df_all["genre"].unique())
print("Detected genres:", genres)

# Optional subset for speed
if FAST_MODE:
    dfs = []
    for g in genres:
        sub = df_all[df_all.genre == g]
        n   = min(FILES_PER_GENRE, len(sub))
        dfs.append(sub.sample(n, random_state=SEID) if (SEID:=SEED) else sub.sample(n, random_state=42))
    df_all = pd.concat(dfs).reset_index(drop=True)
    print(f"FAST_MODE enabled — using {len(df_all)} files total.")

#  Train/Val/Test Split
train_df, temp_df = train_test_split(df_all, test_size=0.3, stratify=df_all["genre"], random_state=SEED)
val_df,   test_df = train_test_split(temp_df, test_size=0.5, stratify=temp_df["genre"], random_state=SEED)
print("Splits:", len(train_df), len(val_df), len(test_df))

# TABULAR (MFCC)
def extract_features_tabular(file_path, sr=SAMPLE_RATE, duration=DURATION, mono=MONO, n_mfcc=N_MFCC):
    try:
        y, sr = librosa.load(file_path, sr=sr, mono=mono, duration=duration)
        if y is None or len(y) < sr:  # shorter than 1 second?
            return None
        # MFCCs
        mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
        mfcc_mean = mfcc.mean(axis=1)
        mfcc_std  = mfcc.std(axis=1)
        # Additional light spectral features
        spec_centroid = librosa.feature.spectral_centroid(y=y, sr=sr).flatten()
        spec_bw       = librosa.feature.spectral_bandwidth(y=y, sr=sr).flatten()
        rolloff       = librosa.feature.spectral_rolloff(y=y, sr=sr).flatten()
        zcr           = librosa.feature.zero_crossing_rate(y).flatten()
        chroma        = librosa.feature.chroma_stft(y=y, sr=sr).mean(axis=1)
        feats = np.hstack([
            mfcc_mean, mfcc_std,
            [spec_centroid.mean(), spec_centroid.std()],
            [spec_bw.mean(), spec_bw.std()],
            [rolloff.mean(), rolloff.std()],
            [zcr.mean(), zcr.std()],
            chroma
        ]).astype(np.float32)
        return feats
    except Exception as e:
        print("Feature error (tabular):", file_path, "->", e)
        return None

def build_tabular_dataset(file_df):
    X_list, y_list = [], []
    for p, g in file_df[["path","genre"]].itertuples(index=False):
        f = extract_features_tabular(p)
        if f is not None:
            X_list.append(f)
            y_list.append(genres.index(g))
    X = np.array(X_list, dtype=np.float32)
    y = np.array(y_list, dtype=np.int64)
    return X, y

tabular_results = {}
if RUN_TABULAR:
    print("\n=== Building TABULAR dataset (MFCC + spectral features) ===")
    X_train, y_train = build_tabular_dataset(train_df)
    X_val,   y_val   = build_tabular_dataset(val_df)
    X_test,  y_test  = build_tabular_dataset(test_df)

    print("Tabular shapes:", X_train.shape, X_val.shape, X_test.shape)

    scaler = StandardScaler()
    X_train_s = scaler.fit_transform(X_train)
    X_val_s   = scaler.transform(X_val)
    X_test_s  = scaler.transform(X_test)

    rf = RandomForestClassifier(
        n_estimators=400,
        random_state=SEED,
        n_jobs=-1
    )
    rf.fit(X_train_s, y_train)

    def eval_tabular(split_name, Xs, ys):
        preds = rf.predict(Xs)
        acc = accuracy_score(ys, preds)
        print(f"[TABULAR] {split_name} Accuracy: {acc:.4f}")
        print(classification_report(ys, preds, target_names=genres, zero_division=0))
        cm = confusion_matrix(ys, preds)
        plt.figure(figsize=(8,6))
        plt.imshow(cm, interpolation="nearest")
        plt.title(f"Tabular Confusion Matrix — {split_name}")
        plt.colorbar()
        ticks = np.arange(len(genres))
        plt.xticks(ticks, genres, rotation=45, ha='right')
        plt.yticks(ticks, genres)
        plt.xlabel("Predicted"); plt.ylabel("True")
        plt.tight_layout(); plt.show()
        return acc

    acc_val_tab  = eval_tabular("Validation", X_val_s, y_val)
    acc_test_tab = eval_tabular("Test", X_test_s, y_test)
    tabular_results = {"val_acc": acc_val_tab, "test_acc": acc_test_tab, "model": rf, "scaler": scaler}

# IMAGE (Mel-Spectrogram + CNN)
def wav_to_melspec_image(path, sr=SAMPLE_RATE, duration=DURATION, n_mels=128, fmin=20, fmax=8000, img_size=IMG_SIZE):
    try:
        y, sr = librosa.load(path, sr=sr, mono=True, duration=duration)
        if y is None or len(y) < sr:  # shorter than 1 second?
            return None
        S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=n_mels, fmin=fmin, fmax=fmax)
        S_db = librosa.power_to_db(S, ref=np.max)  # log-mel

        # Normalize to [0,1]
        S_min, S_max = S_db.min(), S_db.max()
        S_norm = (S_db - S_min) / (S_max - S_min + 1e-9)

        # Resize to (IMG_SIZE, IMG_SIZE)
        img = tf.convert_to_tensor(S_norm[np.newaxis, ..., np.newaxis], dtype=tf.float32)  # (1, H, W, 1)
        img = tf.image.resize(img, (img_size, img_size), method="bilinear")[0]             # (H, W, 1)
        img3 = tf.repeat(img, repeats=3, axis=-1)                                          # (H, W, 3)
        return img3.numpy().astype(np.float32)
    except Exception as e:
        print("Spec error (image):", path, "->", e)
        return None

def build_image_arrays(file_df):
    X_list, y_list = [], []
    for p, g in file_df[["path","genre"]].itertuples(index=False):
        im = wav_to_melspec_image(p)
        if im is not None:
            X_list.append(im)
            y_list.append(genres.index(g))
    X = np.array(X_list, dtype=np.float32)
    y = np.array(y_list, dtype=np.int64)
    return X, y

image_results = {}
if RUN_IMAGE:
    print("\n=== Building IMAGE dataset (mel-spectrograms) ===")
    X_train_i, y_train_i = build_image_arrays(train_df)
    X_val_i,   y_val_i   = build_image_arrays(val_df)
    X_test_i,  y_test_i  = build_image_arrays(test_df)

    print("Image shapes:", X_train_i.shape, X_val_i.shape, X_test_i.shape)

    y_train_cat = to_categorical(y_train_i, num_classes=len(genres))
    y_val_cat   = to_categorical(y_val_i,   num_classes=len(genres))
    y_test_cat  = to_categorical(y_test_i,  num_classes=len(genres))

    inputs = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
    if TRANSFER_LEARNING:
        base = MobileNetV2(include_top=False, input_tensor=inputs, weights="imagenet")
        base.trainable = False  # freeze for a quick, stable baseline
        x = layers.GlobalAveragePooling2D()(base.output)
        x = layers.Dropout(0.3)(x)
        outputs = layers.Dense(len(genres), activation="softmax")(x)
        model = models.Model(inputs=inputs, outputs=outputs)
    else:
        x = layers.Conv2D(32, 3, activation='relu')(inputs)
        x = layers.MaxPool2D()(x)
        x = layers.Conv2D(64, 3, activation='relu')(x)
        x = layers.MaxPool2D()(x)
        x = layers.Conv2D(128, 3, activation='relu')(x)
        x = layers.GlobalAveragePooling2D()(x)
        x = layers.Dropout(0.3)(x)
        outputs = layers.Dense(len(genres), activation='softmax')(x)
        model = models.Model(inputs, outputs)

    model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
    model.summary()

    history = model.fit(
        X_train_i, y_train_cat,
        validation_data=(X_val_i, y_val_cat),
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        verbose=1
    )

    def eval_image(split_name, Xs, ys_int, ys_cat):
        probs = model.predict(Xs, batch_size=BATCH_SIZE, verbose=0)
        preds = probs.argmax(axis=1)
        acc = accuracy_score(ys_int, preds)
        print(f"[IMAGE] {split_name} Accuracy: {acc:.4f}")
        print(classification_report(ys_int, preds, target_names=genres, zero_division=0))
        cm = confusion_matrix(ys_int, preds)
        plt.figure(figsize=(8,6))
        plt.imshow(cm, interpolation="nearest")
        plt.title(f"Image Confusion Matrix — {split_name}")
        plt.colorbar()
        ticks = np.arange(len(genres))
        plt.xticks(ticks, genres, rotation=45, ha='right')
        plt.yticks(ticks, genres)
        plt.xlabel("Predicted"); plt.ylabel("True")
        plt.tight_layout(); plt.show()
        return acc

    acc_val_img  = eval_image("Validation", X_val_i, y_val_i, y_val_cat)
    acc_test_img = eval_image("Test", X_test_i, y_test_i, y_test_cat)
    image_results = {"val_acc": acc_val_img, "test_acc": acc_test_img, "model": model}

#  Summary
print("\n================ Summary ================")
if RUN_TABULAR:
    print(f"Tabular  -> Val Acc: {tabular_results['val_acc']:.4f} | Test Acc: {tabular_results['test_acc']:.4f}")
if RUN_IMAGE:
    print(f"Image(CNN)-> Val Acc: {image_results['val_acc']:.4f} | Test Acc: {image_results['test_acc']:.4f}")
print("=========================================\n")

# Optional: Save to Drive
if ASK_SAVE_TO_DRIVE:
    choice = input("Save trained models to Google Drive? (y/n): ").strip().lower()
    if choice == "y":
        from google.colab import drive
        drive.mount('/content/drive')
        save_dir = "/content/drive/MyDrive/gtzan_models"
        os.makedirs(save_dir, exist_ok=True)
        if RUN_TABULAR:
            joblib.dump(tabular_results["model"], os.path.join(save_dir, "rf_tabular.pkl"))
            joblib.dump(tabular_results["scaler"], os.path.join(save_dir, "scaler_tabular.pkl"))
        if RUN_IMAGE:
            image_results["model"].save(os.path.join(save_dir, "cnn_spectrogram.h5"))
        with open(os.path.join(save_dir, "genres.txt"), "w") as f:
            for g in genres:
                f.write(g + "\n")
        print("Saved to:", save_dir)
    else:
        print("Skipping Drive save.")
```
